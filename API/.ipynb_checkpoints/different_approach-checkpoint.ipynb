```python
import json
import urllib.request
import time
```

```python
# API Wrapper

year = 2022

url = fr"https://ssd.jpl.nasa.gov/api/horizons.api?format=json&COMMAND='301'&OBJ_DATA='YES'&MAKE_EPHEM='YES'&EPHEM_TYPE='OBSERVER'&CENTER='500@399'&START_TIME='{year}-05-26'&STOP_TIME='{year}-07-30'&STEP_SIZE='1h'&QUANTITIES='24"
```

```python
response = urllib.request.urlopen(url)
```

```python
result = json.loads(response.read())
```

```python
data = result['result'].partition('$$SOE')[2].partition('$$EOE')[0]
data
```

```python
times = data.split('\n')[1:-1]
stos = [times[i].split() for i in range(len(times))]

# for time in stos:
#     del time[1]
    
stos
```

```python
from scipy.signal import argrelextrema
import numpy as np

stos_cleaned = np.array(stos)[:,1].astype(float)

new_moon = argrelextrema(stos_cleaned, np.greater)
print(new_moon)
```

```python
from datetime import datetime, timedelta

TWENTYTHREEDAYS = timedelta(days=23)
NINETEENJUNE = datetime(year, 6, 19)

for i in range(len(new_moon[0])):
    tangaroa = datetime.strptime(stos[new_moon[0][i]][0], '%Y-%b-%d') + TWENTYTHREEDAYS
    if tangaroa >= NINETEENJUNE:
        month_start_index = i
        break
        
tangaroa
```

```python
def isTwentyNineDaysMonth():
    month_days = datetime.strptime(stos[new_moon[0][month_start_index+1]][0], '%Y-%b-%d') \
            - datetime.strptime(stos[new_moon[0][month_start_index]][0], '%Y-%b-%d')
    
    print(month_days)
    return True if month_days.days==29 else False

isTwentyNineDaysMonth()
```

```python
def getClosestFriday(d0):
    # If it's a Saturday, the previous Friday is chosen instead of the next
    if (d0.weekday() < 4):
        return d0+timedelta(4-d0.weekday())
    else:
        return d0-timedelta(d0.weekday()-4)
```

```python
matariki = getClosestFriday(tangaroa).strftime('%Y-%m-%d')
matariki
```
